```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("ds_salaries.csv")
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(10,25))
sns.countplot(y=data["job_title"] , data = data , order = data["job_title"].value_counts().index)
plt.xticks(rotation = 45)
plt.show()
```

# What is the average salary for each job title?

```python
data.head(5)
```

```python
# group the data by job title and calculate the average salary
avg_salary_by_job_title = data.groupby('job_title')['salary'].mean()

# convert the values to whole numbers
avg_salary_by_job_title = avg_salary_by_job_title.map(int)

# print the results
print(avg_salary_by_job_title.sort_values(ascending=False))
```

```python
# sort the values in descending order and select the top 10
top_avg_salary = avg_salary_by_job_title.sort_values(ascending=False).head(10)
```

```python
# create a bar chart using seaborn
plt.figure(figsize=(10, 5))
sns.barplot(x=top_avg_salary.values, y=top_avg_salary.index, color='blue')
plt.xlabel('Average Salary')
plt.ylabel('Job Title')
plt.title('Average Salary by Job Title')
plt.show()
```

# How many employees work for each company size category?

```python
plt.figure(figsize=(10,5))
sns.countplot(x=data["company_size"] , data = data)
plt.show()
```

# What is the remote work ratio for each job title?

```python
data.head(5)
```

```python
data_1 = data[['job_title','remote_ratio']]
```

```python
data_1.head()
```

```python
job_remote_grp = data_1.groupby('job_title')['remote_ratio'].mean()
```

```python
# sort the values in descending order and select the top 10
top_job_remote_grp = job_remote_grp.sort_values(ascending=False).head(100)
```

```python
# create a bar chart using seaborn
plt.figure(figsize=(10, 25))
sns.barplot(x=top_job_remote_grp.values, y=top_job_remote_grp.index, color='blue')
plt.xlabel('Remote Ratio')
plt.ylabel('Job Title')
plt.title('Average remote work ratio by Job Title')
plt.show()
```

# Which job titles have the highest and lowest average salaries?

```python
data.head(5)
```

```python
# find the maximum and minimum salary
max_salary = data['salary'].max()
min_salary = data['salary'].min()

# find the corresponding job titles
max_job_title = data.loc[data['salary'] == max_salary, 'job_title'].values[0]
min_job_title = data.loc[data['salary'] == min_salary, 'job_title'].values[0]

# store max and min salary in a tuple
salary_range = (min_salary, max_salary)

# print the result
print(f"The salary range is {salary_range} for the job titles '{min_job_title}' and '{max_job_title}'.")
```

```python
plt.figure(figsize=(10,5))
fig, axes = plt.subplots(1, 2, figsize=(12,6))
sns.barplot(y=[min_salary], x=[min_job_title],ax=axes[0], palette=['blue'])
sns.barplot(y=[max_salary], x=[max_job_title],ax=axes[1], palette=['red'])
plt.show()
```
